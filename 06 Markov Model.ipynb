```python
idx = 0
word2idx = {}
idx2word = {}

corpus = ["I like cat",
          "I love dogs",
          "You hate Animals",
          "You are good"]

def preprocess(text1):
  global idx
  text1 = text1.lower()
  text1 = text1.split()
  alphabets = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

  for i in range(len(text1)):

    word = text1[i]
    for letter in word:
      if letter not in alphabets:
        word=word.replace(letter,'')
    text1[i] = word

    if word not in word2idx:
      word2idx[word] = idx
      idx2word[idx] = word
      idx+=1

  return text1

for i in range(len(corpus)):
  corpus[i]=preprocess(corpus[i])
corpus
```

```python
M = idx
```

```python
import numpy as np

#pi is a M size vector
pi = np.array([float(0)]*M)

def pi_cal(word):
  idx = word2idx[word]
  count = 0

  for docs in corpus:
    if docs[0]==word:
      count+=1

  return (count+1)/(len(corpus)+M)

for word in word2idx:
  pi[word2idx[word]] = pi_cal(word)

# A is a M*M size matrix
A = np.array([[float(0)]*M]*M)

def A_cal(i, j):

  w1 = idx2word[i]
  w2 = idx2word[j]

  countw1 = 0
  for doc in corpus:
    for word in doc:
      if word == w1:
        countw1 += 1

  countw1w2 = 0
  for docs in corpus:
    for i in range(1, len(docs)):
      if docs[i]==w2 and docs[i-1]==w1:
        countw1w2+=1
  
  #print(w1,w2,countw1w2,countw1)
  return (countw1w2+1)/(countw1+M)

for i in range(idx):
  for j in range(idx):
    A[i][j] = A_cal(i,j)  
```

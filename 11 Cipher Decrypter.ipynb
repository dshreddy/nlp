```python
def convert_dna_string_to_map(dna_string):
  map = {}
  alphabets = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p',
       'q','r','s','t','u','v','w','x','y','z']
  for alphabet,ch in zip(alphabets, dna_string):
    map[alphabet] = ch
  return map

def decode(ciphertext, map):
  for key, val in map.items():
    ciphertext  = ciphertext.replace(key, val)
  return ciphertext

def f(dna_string):
  map = convert_dna_string_to_map(dna_string)
  message = decode(ciphertext, map)
  log_likelihood = model(message)
  return log_likelihood
```

```python
def get_random_dna():
  s = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p',
       'q','r','s','t','u','v','w','x','y','z']
  random.shuffle(s)
  return ''.join(s)

def randomly_switch(DNA):

    char_list = list(DNA)
    indices = [i for i in range(26)]

    index1 = random.choice(indices)
    # Making sure that the child is not complete identical to parent
    # By not swapping the same character with itself
    # We can comment this out if we want to
    # But I have never seen a child identical with parent in real life
    indices.remove(index1)
    index2 = random.choice(indices)
    
    char_list[index1], char_list[index2] = char_list[index2], char_list[index1]

    switched_string = ''.join(char_list)    
    return switched_string
```

```python
def get_many_random_dna(num):
  l = []
  t = num
  while t:
    dna = get_random_dna()
    if dna not in l:
      l.append(dna)
      t=t-1
  return l

def create_offspring(DNA_pool, num_childs):
  parents = [dna for dna in DNA_pool]
  for dna in parents:
    # I am not making sure that all 3 childs are unique
    # If we want to we can make sure that the entire next gen kids are diff from prev gen 
    # Or we can also make sure that the 3 childs of parent are unique
    childs = []
    for i in range(num_childs):
      childs.append(randomly_switch(dna))
    for child in childs:
      DNA_pool.append(child)
  return DNA_pool
```

```python
#1st gen DNA's
DNA_pool = get_many_random_dna(20)

for i in range(num_epochs):
  if i >0:
    DNA_pool = create_offspring(DNA_pool, 3)
    scores = [(f(DNA),DNA) for DNA in DNA_pool]
    scores.sort()
    scores.reverse()
    DNA_pool = []
    for i in range(5):
      DNA_pool.append(scores[i][1])
```

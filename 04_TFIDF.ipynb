```python
import pandas as pd
import numpy as np
import nltk

from nltk import word_tokenize
```

```python
nltk.download('punkt')
# necessasry for nltk's word_tokenize
```

```python
df = pd.read_csv("bbc_text_cls.csv")
```

```python
df.head()
```

```python
inputs = df["text"]
labels = df["labels"]
```

```python
idx = 0
word_idx = {}
for doc in inputs:
  tokens = word_tokenize(doc.lower())
  for token in tokens:
    if token not in word_idx:
      word_idx[token] = idx
      word_idx[idx] = token
      idx+=1
print(len(word_idx), idx)
```

```python
N = len(inputs)
V = idx
```

```python
#instantiate term frequency matrix
tf = np.zeros((N,V))

for row in range(N):
  tokens = word_tokenize(inputs[row].lower())
  for token in tokens:
    idx = word_idx[token]
    tf[row][idx]+=1

#compute idf
decoumet_freq = np.sum(tf>0, axis=0)
#look at tf matrix each column at a time, therefore doc_feq is a V size array
idf = np.log(N/decoumet_freq)
```

```python
tf_idf = tf*idf
```

```python
np.random.seed(123)
```

```python
# pick a random document, show the top 5 terms (in terms of tf_idf score)
i = np.random.choice(N)
row = df.iloc[i]
print("Label:", row['labels'])
print("Text:", row['text'].split("\n", 1)[0])
print("Top 5 terms:")

scores = tf_idf[i]
indices = (-scores).argsort()

for j in indices[:5]:
  print(word_idx[j])
```

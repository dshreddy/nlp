```python
#Join all genres, key words into a single string for 1st row
s = ""
for dic in j:
  s+=dic["name"]+","
for dic in jj:
  s+=dic["name"]+","
s
```

```python
#Create training data by repeating what we did above for each row
inputs = []
labels = {}

for row in range(4803):
  x = df.iloc[row]
  j = json.loads(x["genres"])
  jj = json.loads(x["keywords"])
  s = ""
  for dic in j:
    s+=dic["name"]+","
  for dic in jj:
    s+=dic["name"]+","
  inputs.append(s)
  labels[x["title"]]=row
  labels[row]=x["title"]

print(len(inputs), len(labels))
```

```python
def simple_tokenizer(s):
  return s.split(",")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vec = TfidfVectorizer(max_features=2000, tokenizer=simple_tokenizer)
movie_vectors_matrix = vec.fit_transform(inputs)
movie_vectors_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
idx = labels["Scream 3"]
query = movie_vectors_matrix[idx]
scores = []
for row in range(4083):
  scores.append( (-cosine_similarity(query, movie_vectors_matrix[row])[0][0], labels[row]) )
scores.sort()
for i in range(6):
  print(scores[i][1])
```

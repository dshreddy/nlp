```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
```

```python
#creating 100 data points between -3 & +3
N = 100
X = np.random.random(N) * 6 - 3
# y =m*x + c + gaussian noise
y = 0.5 * X - 1 + np.random.randn(N) * 0.5
```

```python
plt.scatter(X,y);
```

```python
# build a model using tensorflow
i = Input(shape=(1,))
x = Dense(1)(i)

model = Model(i,x)
```

```python
model.summary()
```

```python
#defining how our model will be trained

model.compile(
  #mean squared error
  loss='mse',
  # optimizer='adam',
  optimizer=Adam(learning_rate=0.1),
  metrics=['mae']
)
```

```python
r = model.fit(
  X.reshape(-1, 1), y,
  epochs=200,
  batch_size=32,
)
```
